```python
# Customer details table
#---------------------------------------------------------------------------------------------------------------------------
# Open the workbook and define the worksheet
book = xlrd.open_workbook("datasets.xls")
sheet = book.sheet_by_name("CUSTOMER DETAILS")

# Create the INSERT INTO sql query
query = """INSERT INTO customer_details (cust_id, first_name, last_name, registered_date, birth_date, email, address, city, zipcode) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)"""

# Create a For loop to iterate through each row in the XLS file, starting at row 2 to skip the headers
for r in range(1, sheet.nrows):
		cust_id		= sheet.cell(r,0).value
		first_name	= sheet.cell(r,1).value
		last_name	= sheet.cell(r,2).value
		registered_date	= sheet.cell(r,3).value
		birth_date		= sheet.cell(r,4).value
		email	= sheet.cell(r,5).value
		address		= sheet.cell(r,6).value
		city		= sheet.cell(r,7).value
		zipcode		= sheet.cell(r,8).value
        
        
		registered_date = now.strftime('%Y-%m-%d')
		birth_date = now.strftime('%Y-%m-%d')
		# Assign values from each row
		values = (cust_id, first_name, last_name, registered_date, birth_date, email, address, city, zipcode)

		# Execute sql Query
		cursor.execute(query, values)

# Close the cursor
cursor.close()

# Commit the transaction
mydb.commit()

# Print results
print ("")
print ("All Done! Bye, for now.")
columns = str(sheet.ncols)
rows = str(sheet.nrows)
```
